```python
# THIS IS TRAIL PROGRAM!

#COSMOLOGY ASSIGNMENT 6: COSMOLOGY CALCULATOR.

#importing necessary libraries:
import numpy as np
from scipy import integrate

#input of various parameters:
print("PARAMETERS:")
Ho = float(input('Enter the value of Ho (in Km/s/Mpc ):'))
Wm = float(input('Enter the value of \u03A9_\u039C,0 :' ))
Wl = float(input('Enter the value of \u03A9_\u039B,0 :' ))
Wk = float(input('Enter the value of \u03A9_\u039A,0 :' ))
Wr = float(input('Enter the value of \u03A9_\u03B3,0 :' ))
Z  = float(input('Enter the value of Redshift parameter Z \n(Enter only positive values):' ))
c  = 3* 10**(8)       # in m sec^ -1


def calc(Z,Wm,Wl,Wk,Wr,Ho):
    #1.Present age of universe (in Gyears)
    #2.Age of the universe for user input redshift (in Gyears)
    #3.Look back time for user input redshift (in Gyears)
    
    def TIME(Z):
        return ( 1/( (1+Z)*(((Wm*((1+Z)**3)) + (Wr*((1+Z)**4)) + Wl + (Wk*((1+Z)**2)))**(1/2)) ) )
    t0,err = integrate.quad(TIME,0,np.inf)
    T0     = t0*( 3.08*(10**(19)) )/(Ho*3.15 * (10**(16)))            #in Giga years
    tz,err = integrate.quad(TIME,Z,100000)
    Tz     = tz*( 3.08*(10**(19)) )/(Ho*3.15 * (10**(16)))            #in Giga years
    del_T  = (T0 - Tz)
    print("1.Present age of universe (in Gyears):",np.round(T0,3))
    print("2.Age of the universe for user input redshift (in Gyears):",np.round(Tz,3))
    print("3.Look back time for user input redshift (in Gyears):",np.round(del_T,3))
    
    #4.Comoving radial distance to user input redshift (in MPC)
    #5.Angular diameter distance to user input redshift (in MPC)
    #6.Angular scale at user input redshift (kpc/arcsec)
    #7.Luminosity distance to user input redshift (in MPC)

    def fun(Z,Wm,Wl,Wk,Wr):
        return(1*( 3.08*(10**(19)) )/(((Wm*((1+Z)**3)) + (Wr*((1+Z)**4)) + Wl + (Wk*((1+Z)**2)))**(1/2)))

    r,err = integrate.quad(fun,0,Z,args=(Wm,Wl,Wk,Wr))
    rc = r*c/(Ho*3.08* 10**(22))              #(in MPC)
    dl = (r*c*(1+Z))/(Ho*3.08* 10**(22))      #(in MPC)
    da = (r*c)/(Ho*(1+Z)*(3.08* 10**(22)))    #(in MPC)
    ang = (da*(10**3))/(206265)

    print("4.Comoving radial distance to user input redshift (in MPC)",np.round(rc,3))
    print("5.Angular diameter distance to user input redshift (in MPC)",np.round(da,3))
    print("6.Angular scale at user input redshift (kpc/arcsec)",np.round(ang,3))
    print("7.Luminosity distance to user input redshift (in MPC)",np.round(dl,3))
    
    #8. Temperature of CMB at user input redshift(K):

    Temp0 = 2.7
    Tcmb  = Temp0*(1+Z) #in Kelvin
    print("8.Temperature of CMB at user input redshift(K):",np.round(Tcmb,3))

print("\n\n RESULTS:")
calc(Z,Wm,Wl,Wk,Wr,Ho)
```
